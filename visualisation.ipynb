```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
df.head()
```

```python
nombre_lignes, nombre_colonnes = df.shape
print(f"Nombre de lignes : {nombre_lignes}")
print(f"Nombre de colonnes : {nombre_colonnes}")
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' contains the Women's E-Commerce Clothing Reviews dataset

# Histogram of ratings using Seaborn
sns.histplot(df['Rating'], bins=5, color='skyblue', kde=False)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings')
plt.show()

# Scatter plot of 'Age' vs. 'Positive Feedback Count' using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Positive Feedback Count', hue='Rating', data=df, palette='viridis')
plt.xlabel('Age')
plt.ylabel('Positive Feedback Count')
plt.title('Scatter Plot of Age vs. Positive Feedback Count')
plt.show()

# Box plot of 'Positive Feedback Count' grouped by 'Rating' using Seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x='Positive Feedback Count', y='Rating', data=df, orient='h')
plt.xlabel('Positive Feedback Count')
plt.ylabel('Rating')
plt.title('Box Plot of Positive Feedback Count Grouped by Rating')
plt.show()

# Line graph showing the average 'Age' for each 'Rating' level using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='Rating', y='Age', data=df, marker='o', err_style=None)

plt.xlabel('Rating')
plt.ylabel('Average Age')
plt.title('Average Age by Rating')
plt.show()

# Pie chart showing the distribution of 'Department Name' categories using Matplotlib
plt.figure(figsize=(8, 8))
dept_counts = df['Department Name'].value_counts()
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%', colors=sns.color_palette('tab20'))
plt.title('Distribution of Department Name Categories')
plt.show()

# Creating a table to display summary statistics by 'Rating'
summary_table = df.groupby('Rating').agg({
    'Age': 'mean',
    'Positive Feedback Count': 'mean'
}).reset_index()

print(summary_table)
```

Most of the rating are 3-5 which describe product quality are good
Most of the women recommended the products
clearly shows that good rating means product will be recommended
